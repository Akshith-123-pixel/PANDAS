# INTORDUCTION TO SEABORN 

seaborn is very powerfull visulization(statistical visulaization) pkg in python

```python
import matplotlib.pyplot as plt     # visualization 
import seaborn as sns   # distribution visulaization 
# sea born is usefull for advance visualization ex -- distirbution plot , line plot
%matplotlib inline
plt.rcParams['figure.figsize'] = 9,5
import warnings
warnings.filterwarnings('ignore') # os error
```

```python
stats.head()
```

```python
stats['InternetUsers']
```

# Distributions

```python
vis1 = sns.distplot(stats["InternetUsers"])   # univariant analysis
```

```python
vis1
plt.show()
```

```python
vis1 = sns.distplot(stats['InternetUsers'] , bins = 10)
vis1
plt.show()  # here bin is use for better visualization 
```

# box plot

```python
vis2 = sns.boxplot(data = stats , x = 'IncomeGroup', y = 'BirthRate')  # Bi- variate analyisis
vis2
plt.show()
```

in this graph circles are called as outliers , max and min points are there , down one is 25% and upperone is 75% and in the miidle line is average line 
Low income has more avg , High income as less avg 

```python
vis4 = sns.lmplot(data = stats , x = 'InternetUsers' , y = 'BirthRate')
vis4 
plt.show()  # lmplot is the function of seaborn library (linear model)
```

in this graph these all points are dataframe points of 195 these are plot as according to birthrate and internet users 

```python
vis3 = sns.lmplot(data = stats , x = 'InternetUsers', y = 'BirthRate' , fit_reg = True)
vis3
plt.show()
```

```python
vis2 = sns.lmplot(data = stats , x = 'InternetUsers', y = 'BirthRate' , fit_reg = False)
```

```python
vis2
plt.show()
```

 in this graph there is no middle line we mention as fit_reg = False that's why it result as no middle line if we given true the it return as middle line 

```python
vis1 = sns.lmplot(data = stats , x = 'InternetUsers' , y = 'BirthRate' , fit_reg = False , hue = 'IncomeGroup')
vis1 
plt.show()
```

if we use hue is for colour parameter and this graph is for better understanding to the client (low income has high birthrate and high invome has low birthrate.)

```python
vis5 = sns.lmplot(data=stats , x ='InternetUsers' , y = 'BirthRate' , fit_reg = True , hue = 'IncomeGroup' , height = 7)
vis5
plt.show()
```

in this above graph i mentioned the height 7
